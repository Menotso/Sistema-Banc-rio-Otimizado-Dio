```python
saldo = 0
saques = 0
extrato_depositos = []
extrato_saques = []
cadastros = {}

def menu_cadastro():
    global saldo
    global cadastros
    print('Bem-vindo ao Banco DIO!')
    print('Digite (0) para fazer um novo cadastro;')
    print('Digite (1) caso possua um cadastro')
    print('Digite (6) para encerrar.')
    opcao_escolhida = input('Insira um número: ')
    if opcao_escolhida == '0':
        cadastros.update(novo_cadastro())
        menu()
        print('Cadastro realizado!!!\nRetornando ao MENU.')
    elif opcao_escolhida == '1':
        verifica_cadastro()
    elif opcao_escolhida == '6':
        print('Encerrando...')
        return f'Programa encerrado.'
    else:
        print('Retornando ao menu virtude opção escolhida ser INVÁLIDA.')
        menu_cadastro()
    
        
    
def menu():
    print('Digite (2) para fazer um depósito;')
    print('Digite (3) para efetuar um saque;')
    print('Digite (4) para retirar o extrato;')
    print('Digite (5) para exibir a lista de cadastros;')
    print('Digite (6) para sair.')
    menu_cumpre_opcoes()
        
def menu_cumpre_opcoes():
        global opcao_escolhida
        opcao_escolhida = input('Insira um número: ')
        if opcao_escolhida == '2':
            deposito()
        elif opcao_escolhida == '3':
            sacar()
        elif opcao_escolhida == '4':
            global extrato_escolhido
            extrato_escolhido = str(input('Escolha uma das opções a seguir: \nsaques, depositos, saldo, extrato geral'))
            extrato()
        elif opcao_escolhida == '5':
            print(f'LISTA DE CADASTROS:')
            print(cadastros)
        elif opcao_escolhida == '6':
            print('Encerrando...')
            return f'Programa encerrado.'
        else:
            print('Serviço inexistente, voltando ao menu.')
            return menu()

def novo_cadastro():
    global cadastros
    print('Bem vindo à seção de cadastro !')
    cria_cpf = str(input('Insira seu CPF'))
    if cria_cpf in cadastros:
        print('Este CPF já consta no banco de dados, não foi possível criar novo cadastro.\n Retornando ao Menu.')
        return menu()
    formata_novo_cpf = deletar_caractere_repetido(string=cria_cpf, caractere=".")
    cria_nome_usuario = str(input('Insira seu usuário: '))
    cria_conta_corrente = str(input('Insira agora o seu número da conta corrente'))
    cria_data_nascimento = str(input('Insira sua data de nascimento'))
    cria_logradouro = str(input('Insira na seguinte ordem, o seu endereço: logradouro, nro - bairro - cidade/sigla estado'))
    novo_usuario = {formata_novo_cpf : [
        cria_nome_usuario,
        cria_data_nascimento,
        cria_conta_corrente,
        cria_logradouro
    ]}
    return novo_usuario

def verifica_cadastro():
    cadastro_a_ser_verificado = str(input('Insira seu CPF'))
    cpf_formatado = deletar_caractere_repetido(string=cadastro_a_ser_verificado, caractere=".")
    if cpf_formatado in cadastros:
        return menu
    else:
        print("Cadastro não encontrado, por favor tente novamente ou registre-se.")
        return menu_cadastro()
        
def deletar_caractere_repetido(*, string='', caractere=''):
    nova_string = ""
    for char in string:
        if char != caractere:
            nova_string += char
    return nova_string

def deposito():
    global saldo
    while True:
        try:
            deposito = int(input('Qual valor deseja depositar? '))
            if deposito <= 0:
                print('O valor do depósito deve ser maior que zero.')
            else:
                saldo += deposito
                print('Seu novo depósito foi registrado com sucesso!')
                break
        except ValueError:
            print('Valor inválido. Digite um número inteiro.')
    global extrato_depositos
    extrato_depositos.append(deposito)
    menu()

def sacar():
    global saldo
    global saques
    sacar = int(input('Digite a quantidade do saque: '))
    if saques >= 4:
        print('Limite de Saques diário atingido. Aguarde até o dia seguinte.')
        return menu()
    elif saldo <= sacar:
        print('Não há saldo suficiente para esta operação.\nRetornando ao Menu.')
        return menu()
    elif sacar > 500:
        print('Saque maior do que o limite autorizado.\nRetornando ao Menu.')
        return menu()
    saques += 1
    saldo -= sacar
    print(f'Saque concluído, saque {saques} de 3.\nRetornando ao Menu.')
    global extrato_saques
    extrato_saques.append(sacar)
    menu()
    
def extrato(relatorio_saques='saques', relatorio_depositos='depositos', relatorio_saldo='saldo', relatorio_geral='extrato geral'):
    global extrato_escolhido
    if extrato_escolhido == relatorio_saques:
        return print(f'Relatório de Saques: {extrato_saques}'), menu()
    elif extrato_escolhido == relatorio_depositos:
        return print(f'Relatório de Depositos: {extrato_depositos}'), menu()
    elif extrato_escolhido == relatorio_saldo:
        return print(f'Seu saldo atual: R${saldo}'), menu()
    elif extrato_escolhido == relatorio_geral:
        return print(f"""
            Relatório de Saques: {extrato_saques}
            Relatório de Depositos: {extrato_depositos}
            Seu saldo atual: R${saldo}
        """), menu()
    else:
        print('Opção não reconhecida, por favor tente novamente')
        return menu()
        
if __name__ == "__main__":
    menu_cadastro()
```
